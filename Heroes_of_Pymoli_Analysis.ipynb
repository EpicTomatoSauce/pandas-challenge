```python
#import pandas
import pandas as pd
```

```python
#create reference to csv file
csv_path = "Resources/purchase_data.csv"

#and import as dataframe
heroes = pd.read_csv(csv_path, low_memory = False)
heroes.head(5)
```

```python
#determine the total number of players. NB that as this is purchase data, purchaser SN may habe repeated values - count unique?
unique_player_count = len(pd.unique(heroes['SN']))
#output count of 576 players.
#create new dataframe to display count of unique_player_count as Total Players
total_players_df = pd.DataFrame({'Total Players':[unique_player_count]})
total_players_df
```

```python
# Purchasing Analysis (Total)
# Run basic calculations to obtain number of unique items, average price, etc.
# Create a summary data frame to hold the results
# Optional: give the displayed data cleaner formatting
# Display the summary data frame

#get values
num_unique_items = len(pd.unique(heroes['Item Name']))
average_purch_price = heroes['Price'].mean()
num_purchases = heroes['Purchase ID'].count()
total_revenue = heroes['Price'].sum()

#create dataframe with appropriate formatting
purchasing_analysis_df = pd.DataFrame({'Number of Unique Items':[num_unique_items],
                                      'Average Purchase Price':[average_purch_price],
                                      'Number of Purchases':[num_purchases],
                                      'Total Revenue':[total_revenue]}).style.format({'Average Purchase Price':"${:,.2f}",
                                                                                     'Total Revenue':"${:,.2f}"})


purchasing_analysis_df
                          
```

```python
# Gender Demographics
# Percentage and Count of Male Players
# Percentage and Count of Female Players
# Percentage and Count of Other / Non-Disclosed

#in order to create table summarising Gender, create groupby object. N.B same as above, may need ot consider unique values.
#groupby on initial import
genders = heroes.groupby("Gender")

#using categories defined by the groupby, count unique values in SN column
total_count_gender = genders.nunique()["SN"] 

#total_count_gender #result consistent with sample file

#calculate percentages using unique_player_count
percentage_genders = total_count_gender / unique_player_count * 100

#create dataFrame
gender_demographics_df = pd.DataFrame({"Total Count":total_count_gender,
                                      "Percentage of Players":percentage_genders})

#sort and convert percentage column
gender_demographics_sorted_df = gender_demographics_df.sort_values(by=["Total Count"], ascending = False).style.format({"Percentage of Players":"{:,.2f}%"})

gender_demographics_sorted_df
```

```python
# Purchasing Analysis (Gender)
# Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender
# Create a summary data frame to hold the results
# Optional: give the displayed data cleaner formatting
# Display the summary data frame

#gender categorisation already executed in [6]
purchase_count_gender = genders["Purchase ID"].count()
avg_purchase_value_gender = genders["Price"].mean()
total_purchase_value_gender = genders["Price"].sum()
#average total purchase per person requires count of unique player IDs from above
avg_total_purchase_per_person_gender = total_purchase_value_gender / total_count_gender


#create dataframe
purchasing_analysis_gender_df = pd.DataFrame({"Purchase Count":purchase_count_gender,
                                             "Average Purchase Price":avg_purchase_value_gender,
                                             "Total Purchase Value":total_purchase_value_gender,
                                             "Average Total Purchase per Person":avg_total_purchase_per_person_gender})

#sort and style dataframe
purchasing_analysis_gender_sorted_df = purchasing_analysis_gender_df.sort_values(by=["Purchase Count"], ascending = False)\
.style.format({"Average Purchase Price":"${:,.2f}",
               "Total Purchase Value":"${:,.2f}",
               "Average Total Purchase per Person":"${:,.2f}"})

purchasing_analysis_gender_sorted_df
```

```python
# Age Demographics
# Establish bins for ages
# Categorize the existing players using the age bins. Hint: use pd.cut()
# Calculate the numbers and percentages by age group
# Create a summary data frame to hold the results
# Optional: round the percentage column to two decimal points
# Display Age Demographics Table

#show max and min as a reference
# max_age = heroes['Age'].max()
# min_age = heroes['Age'].min()
# print(f'{max_age} and {min_age}')
#Max of 45 and Min of 7

#Create bins and groups
age_bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 999]
age_groupings = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

#Create new column and assign bin
heroes["Age Group"] = pd.cut(heroes["Age"], age_bins, labels = age_groupings, include_lowest = True)
age_groupings = heroes.groupby("Age Group")

#count unique player IDs within Age Group and calculate percentages based on unique_player_count
total_count_age = age_groupings["SN"].nunique()
percentage_players_age = total_count_age / unique_player_count * 100

#create dataframe
age_demographics_analysis_df = pd.DataFrame({"Total Count":total_count_age,
                                             "% of Players":percentage_players_age})

age_demographics_analysis_sorted_df = age_demographics_analysis_df.style.format({"% of Players":"{:,.2f}%"})

age_demographics_analysis_sorted_df
```

```python
# Purchasing Analysis (Age)
# Bin the purchase_data data frame by age
# Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below
# Create a summary data frame to hold the results
# Optional: give the displayed data cleaner formatting
# Display the summary data frame
# uses age binning from above (age_groupings)

purchase_count_age = age_groupings["Purchase ID"].count()
avg_purchase_price_age = age_groupings["Price"].mean()
total_purchase_value_age = age_groupings["Price"].sum()
avg_total_purchase_per_age = total_purchase_value_age / total_count_age

age_purchasing_analysis_df = pd.DataFrame({"Purchase Count":purchase_count_age,
                                           "Average Purchase Price":avg_purchase_price_age,
                                           "Total Purchase Value":total_purchase_value_age,
                                           "Average Total Purchase per Person":avg_total_purchase_per_age})\
.style.format({"Average Purchase Price":"${:,.2f}",
               "Total Purchase Value":"${:,.2f}",
               "Average Total Purchase per Person":"${:,.2f}"})

age_purchasing_analysis_df
```

```python
# Top Spenders
# Run basic calculations to obtain the results in the table below
# Create a summary data frame to hold the results
# Sort the total purchase value column in descending order
# Optional: give the displayed data cleaner formatting
# Display a preview of the summary data frame

#need to groupby on the SN to identfy transactions by purchaser
spenders = heroes.groupby("SN")
purchase_count_spenders = spenders["Purchase ID"].count()
avg_purchase_price_spenders = spenders["Price"].mean()
total_purchase_value_spenders = spenders["Price"].sum()

top_spenders_df = pd.DataFrame({"Purchase ID":purchase_count_spenders,
                               "Average Purchase Price":avg_purchase_price_spenders,
                               "Total Purchase Value":total_purchase_value_spenders})

top_spenders_sorted_df = top_spenders_df.sort_values(by=["Purchase ID"], ascending = False)

top_spenders_sorted_df.head(5).style.format({"Average Purchase Price":"${:,.2f}",
                                     "Total Purchase Value":"${:,.2f}"})
```

```python
# Most Popular Items
# Retrieve the Item ID, Item Name, and Item Price columns
# Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value
# Create a summary data frame to hold the results
# Sort the purchase count column in descending order
# Optional: give the displayed data cleaner formatting
# Display a preview of the summary data frame

#create new dataframe from original with the three categories and then group by Item ID
items_df = heroes[["Item ID", "Item Name", "Price"]]
grouped_items_df = items_df.groupby(["Item ID", "Item Name"])
purchase_count = grouped_items_df["Price"].count()
total_item_purchase = grouped_items_df["Price"].sum()
item_price = total_item_purchase / purchase_count

item_popularity_df = pd.DataFrame({"Purchase Count":purchase_count, 
                                   "Item Price": item_price, 
                                   "Total Purchase Value": total_item_purchase})

item_popularity_sorted_df = item_popularity_df.sort_values(by=["Purchase Count"], ascending = False)

item_popularity_sorted_df.head(5).style.format({"Item Price":"${:,.2f}",
                                               "Total Purchase Value":"${:,.2f}"})
   
```

```python
# Most Profitable Items
# Sort the above table by total purchase value in descending order
# Optional: give the displayed data cleaner formatting
# Display a preview of the data frame

#re-sort the original item_popularity_df in descending order on total purchase value
most_profitable_df = item_popularity_df.sort_values(by=["Total Purchase Value"], ascending = False)
most_profitable_df.head(5).style.format({"Item Price":"${:,.2f}",
                                               "Total Purchase Value":"${:,.2f}"})
```
